# Value-at-Risk from a Pragmatic Trading Perspective

Value-at-Risk (VaR) is often introduced as a statistical quantile of a loss distribution, but in trading it is more naturally understood as a daily risk-control number attached to a marked-to-market portfolio.

In this notebook, we discuss VaR from a pragmatic point of view: what object it is applied to, how portfolio value is measured, how losses and returns enter the computation, and how VaR is usually updated through time.

## Portfolio Value and Mark-to-Market

A trading portfolio has a value at each time $t$, denoted by

$$
V_t.
$$

In liquid markets, this value is usually marked-to-market. This means that the portfolio is valued using current market prices, not original purchase prices. If the portfolio contains positions $w_{i,t}$ in assets with prices $P_{i,t}$, then a simple marked-to-market portfolio value is

$$
V_t
=
\sum_{i=1}^n w_{i,t}P_{i,t}
+
C_t,
$$

where $C_t$ is cash.

In practice, the sequence

$$
V_0,V_1,V_2,\ldots
$$

is the portfolio equity curve. For daily VaR, this equity curve is usually observed or reconstructed at daily frequency, although intraday VaR is also possible.

The daily portfolio profit and loss is

$$
\Delta V_t
=
V_t - V_{t-1}.
$$

The corresponding portfolio return is

$$
R_t
=
\frac{V_t - V_{t-1}}{V_{t-1}}.
$$

When losses are preferred as positive quantities, we define

$$
L_t
=
-\Delta V_t
$$

for dollar losses, or

$$
\ell_t
=
-R_t
$$

for percentage losses.

Thus, VaR can be expressed either in portfolio-value units or in return units.

## VaR in Return Units

When VaR is computed from portfolio returns, it estimates a loss percentage. For confidence level $q$, the return-based VaR is a number $\operatorname{VaR}^{(R)}_q$ such that

$$
\Pr(\ell_t > \operatorname{VaR}^{(R)}_q)
=
1-q.
$$

For example, if the one-day 99% return VaR is $2\%$, then the interpretation is:

> Based on the model, the portfolio is expected to lose more than 2% in one day about 1% of the time.

This is scale-free and useful for comparing risk across portfolios of different sizes.

## VaR in Portfolio-Value Units

Trading desks and risk reports often need VaR in money units. If the portfolio value at time $t$ is $V_t$, then a return-based VaR can be converted into money VaR by

$$
\operatorname{VaR}^{(\$)}_{q,t+1}
=
V_t\cdot \operatorname{VaR}^{(R)}_{q,t+1}.
$$

For example, if the portfolio is worth USD 10 million and the one-day 99% return VaR is $2\%$, then

$$
\operatorname{VaR}^{(\$)}
=
10{,}000{,}000 \times 0.02
=
200{,}000.
$$

So the portfolio has a one-day 99% VaR of USD 200,000.

This is why VaR is naturally tied to the current marked-to-market portfolio value. As the equity curve changes, the same percentage VaR corresponds to a different dollar VaR.

## Portfolio-Level VaR versus Asset-Level VaR

VaR should usually be computed at the portfolio level, not merely asset by asset. The reason is that portfolio risk depends on how the assets move together.

If the portfolio return is

$$
R_{p,t}
=
\sum_{i=1}^n a_{i,t}R_{i,t},
$$

where $a_{i,t}$ are portfolio weights, then VaR should be computed from the portfolio return series

$$
R_{p,t}.
$$

This automatically incorporates the effect of position sizes, long and short exposures, and historical co-movement among assets.

A common mistake is to compute VaR separately for each asset and then add those VaRs together. This usually overstates or distorts portfolio risk because it ignores diversification and correlation structure.

## Fixed Historical VaR versus Rolling VaR

VaR can be computed using one fixed historical sample or using a rolling window.

A fixed historical VaR uses one sample period, such as all available returns from 2003 to 2026, and estimates one VaR number from that entire sample. This is simple, but it assumes that the full historical period is equally relevant to the current portfolio risk.

A rolling VaR updates through time. For each date $t$, it uses only the most recent $\ell$ observations,

$$
R_{t-\ell},R_{t-\ell+1},\ldots,R_{t-1},
$$

to estimate the VaR for the next day. The window then moves forward one day at a time.

In trading, rolling VaR is usually more natural because market volatility changes. A VaR estimate from a calm period should not be treated the same as a VaR estimate during a crisis period.

The choice of rolling window length is therefore important. A short window reacts quickly but may be noisy. A long window is more stable but may react slowly to regime changes.

## One-Day-Ahead VaR

The most common trading interpretation is one-day-ahead VaR. At the end of day $t$, using information available up to day $t$, the model estimates the loss threshold for day $t+1$:

$$
\widehat{\operatorname{VaR}}_{q,t+1}.
$$

The next day, the realized loss $L_{t+1}$ is observed. A VaR violation occurs if

$$
L_{t+1}
>
\widehat{\operatorname{VaR}}_{q,t+1}.
$$

This makes VaR a forecasting object, not merely a descriptive statistic. It must be judged by how well yesterday's forecast compares with today's realized loss.

## How VaR Is Used in Trading

In a trading context, VaR is mainly used as a risk-budgeting and monitoring tool.

If a portfolio has current value $V_t$ and one-day 99% VaR of USD 200,000, then the trader or risk manager knows that the model regards a one-day loss beyond USD 200,000 as a 1% tail event.

This can be used to control exposure. For example, if the allowed daily VaR limit is USD 150,000 but the current portfolio VaR is USD 200,000, then the portfolio is above its risk budget. The trader may need to reduce position sizes, hedge exposures, or avoid adding further risk.

VaR is therefore not just a number describing the past. It is often connected to position sizing, leverage control, stop-review procedures, and risk-limit governance.

## VaR and the Equity Curve

Because the portfolio is marked-to-market, VaR interacts directly with the equity curve.

Suppose a portfolio loses money and its value falls from $V_t$ to $V_{t+1}$. Even if the percentage VaR stays the same, the dollar VaR changes because the capital base has changed. Conversely, if volatility rises, the percentage VaR can increase even if the portfolio value is unchanged.

Thus, dollar VaR is affected by two forces:

$$
\operatorname{VaR}^{(\$)}_{q,t}
=
V_t\cdot \operatorname{VaR}^{(R)}_{q,t}.
$$

The first component, $V_t$, is the current marked-to-market portfolio value. The second component, $\operatorname{VaR}^{(R)}_{q,t}$, is the estimated percentage loss threshold.

This is why VaR should be viewed together with the equity curve, not separately from it.

## VaR as a Model-Dependent Risk Estimate

VaR is not observed directly. It is estimated from a model.

Different models can produce different VaR forecasts from the same portfolio return series. A normal VaR model may give one estimate, historical simulation another, and a GARCH or EVT model another.

This means VaR is only as good as its assumptions. In practice, we therefore care not only about the VaR number itself, but also whether the VaR model is properly backtested.

A good VaR model should produce violations at approximately the expected rate. For example, a 99% VaR model should be violated about 1% of the time. The violations should also not be excessively clustered, because clustered violations indicate that the model is failing precisely when risk is changing.

## Practical Summary

From a pragmatic trading perspective, VaR is best understood as a rolling, model-based estimate of potential portfolio loss.

The usual workflow is:

1. Mark the portfolio to market and obtain the equity curve.
2. Compute portfolio returns or portfolio losses.
3. Estimate one-day-ahead VaR using a chosen model and rolling window.
4. Convert return VaR into dollar VaR using current portfolio value when needed.
5. Compare realized losses against VaR forecasts.
6. Backtest the model by studying violation frequency and violation clustering.

Thus, VaR is not merely a theoretical quantile. It is a practical risk-control object attached to a changing marked-to-market portfolio.

## Project Scope: VaR and CVaR Methods

This subfolder is devoted to Value-at-Risk (VaR), Conditional Value-at-Risk (CVaR), and related risk-measurement methods. The list of notebooks is intended to be expandable. We begin with the most basic VaR constructions, then gradually move toward more model-dependent methods such as filtered historical simulation, GARCH-based VaR, EVT-based VaR, and GARCH-GPD conditional EVT.

The guiding idea is simple: VaR is not a single model. VaR is a risk quantity, and different models estimate it in different ways. Some methods use the historical portfolio loss distribution directly. Others impose distributional assumptions. More advanced methods try to model time-varying volatility, tail behavior, or both.
